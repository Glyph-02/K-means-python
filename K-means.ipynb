```python
from sklearn.datasets import make_moons, make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basic alghoritm k-means
X = np.array([[1,2], [2,1], [2,2], [5,6], [5,5],[6,5],[6,6],[7,5],[7,6],[5,7],[6,7], [1,3],[3,2],[3,3],[2,3],[3,1]])
kmeans = KMeans(n_clusters=2).fit(X)
```

```python
#Assigning data to clusters
kmeans.labels_
```

```python
#Coordinates of clusters centers
kmeans.cluster_centers_
```

```python
plt.scatter (X[:,0],X[:,1], c=kmeans.labels_)
```

```python
#Another example with bigger dataset
X, y = make_blobs(n_samples=400, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50, color = 'grey')
plt.title("Data before assigned to clusters")
```

```python
ykmeans= KMeans(n_clusters=4).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=ykmeans, s=50, cmap='Set2');
plt.title("Data assigned to clusters")
```

```python
n_samples = 400
random_state = 170
X, y = make_moons(n_samples=n_samples, random_state=random_state, noise = 0.1)
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=plt.cm.Set2)
plt.title("Incorrect Shape of data")
```
